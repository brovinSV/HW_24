```python
import random
import string
import time
```

```python
int_list = []
float_list = []
str_list = []
```

```python
# Заповнюємо int_list випадковими цілочисельними значеннями
for i in range(0, 5000):
    int_list.append(random.randint(0, 1000))
#print(int_list)
```

```python
# Заповнюємо float_list випадковими цілочисельними значеннями
for i in range(0, 5000):
    float_list.append(random.uniform(0.1, 100.0))
#print(float_list)
```

```python
# Заповнюємо словник str_list випадковими словами

for i in range(0, 5000):
    word = ''.join(random.choice(string.ascii_letters) for i in range(5))
    str_list.append(word)
#print(str_list)    
```

```python
def bubble_sort_run_time(my_list):
    cur_time = time.time()
    for run in range(len(my_list)-1):
        for i in range(len(my_list)-1-run):
            if my_list[i] > my_list[i + 1]:
                my_list[i], my_list[i +1] = my_list[i + 1], my_list[i]
    print(my_list)
```

```python
%timeit bubble_sort_run_time(str_list)
```

```python
%timeit bubble_sort_run_time(int_list)
```

```python
%timeit bubble_sort_run_time(float_list)
```
