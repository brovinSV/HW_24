```python
"""
1. Два економісти заповнюють документи з власними прогнозами щодо курсу певних коштовних паперів, після чого їх складають 
у спільну папку. Ймовірність відхилення прогнозованого значення курсу від фактичного понад допустиму похибку для першого 
економіста дорівнює 0,1; для другого - 0,2. Перший економіст заповнив 40 документів, другий - 60. Під час перевірки навмання 
взятий із папки документ виявився з суттєвою похибкою. Визначити ймовірність того, що його склав перший економіст.
"""
```

```python
# Задані умови
p_error_A = 0.1  # ймовірність похибки у першого економіста
p_error_B = 0.2  # ймовірність похибки у другого економіста
n_A = 40  # кількість документів першого економіста
n_B = 60  # кількість документів другого економіста
```

```python
"""
Для вирішення завдання застосуємо формулу Байєса:

P(A|B) = P(B|A) * P(A) / P(B),

де
P(A|B) - ймовірність того, що документ складено першим економістом, за умови, що він містить суттєву похибку;
P(B|A) - ймовірність того, що документ містить суттєву похибку, за умови, що він складений першим економістом (0.1);
P(A) - апріорна ймовірність того, що документ складено першим економістом (ймовірність рівномірного розподілу між 
економістами, тобто 0.5);
P(B) – ймовірність того, що документ містить суттєву похибку.
"""

# Розрахунок ймовірності помилки у документі
p_error = p_error_A * n_A / (p_error_A * n_A + p_error_B * n_B)
```

```python
# Виведення результата
print(f"Імовірність того, що документ складено першим економістом, за умови, що він містить суттєву похибку: {p_error}")
```
